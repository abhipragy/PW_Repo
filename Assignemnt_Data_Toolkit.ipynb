4.  Explain the use of the groupby() method in Pandas.

    The groupby() method in Pandas is used to group rows in a DataFrame based on the values of one or more columns. This allows for the application of aggregate functions (e.g., sum, mean, count) to each group independently. It follows a "split-apply-combine" strategy, where the data is split into groups, a function is applied to each group, and the results are combined.


```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 20, 15, 25, 12]}
df = pd.DataFrame(data)

# Group by 'Category' and calculate the sum of 'Value' for each group
grouped_sum = df.groupby('Category')['Value'].sum()
print(grouped_sum)
# Output
# Category
# A    37
# B    45
# Name: Value, dtype: int64

# Group by 'Category' and calculate the mean of 'Value' for each group
grouped_mean = df.groupby('Category')['Value'].mean()
print(grouped_mean)
# Output
# Category
# A    12.333333
# B    22.500000
# Name: Value, dtype: float64
```

6.   What are the differences between NumPy arrays and Python lists

    NumPy arrays and Python lists are both used to store collections of data, but they have key differences that make them suitable for different purposes.

    Data Types:
    NumPy arrays can only store elements of the same data type, while Python lists can store elements of different data types. This homogeneity allows NumPy to perform operations more efficiently.

    Memory Efficiency:
    NumPy arrays are more memory-efficient than Python lists, especially when dealing with large datasets. NumPy stores elements in a contiguous block of memory, while Python lists store elements as separate objects scattered in memory.

    Performance:
    NumPy is optimized for numerical computations and offers significantly faster performance than Python lists for operations like element-wise addition, multiplication, and other mathematical functions. These operations are implemented in C, which contributes to the speed advantage.
   
    Size:

    NumPy arrays have a fixed size, while Python lists are dynamic and can be resized. This means that adding or removing elements from a NumPy array can be less efficient than doing the same with a Python list.
   
    Operations:
    NumPy provides a wide range of built-in functions for array manipulation, linear algebra, and other mathematical operations. Python lists have fewer built-in operations and often require loops for similar tasks.
    
    Flexibility:
    Python lists are more flexible than NumPy arrays because they can store different data types and can be easily modified. NumPy arrays are more rigid but offer better performance for numerical computations.

20.   What is the role of a pivot table in Pandas.

    A pivot table in Pandas is a way to reshape and summarize data within a DataFrame. It allows for rearranging data from a "long" format to a "wide" format, making it easier to analyze and extract insights. It is particularly useful for aggregating data based on one or more index columns and one or more column values


```python
import pandas as pd
import numpy as np

# Sample DataFrame
data = {'Date': ['2025-01-01', '2025-01-01', '2025-01-02', '2025-01-02', '2025-01-01', '2025-01-02'],
        'City': ['London', 'Paris', 'London', 'Paris', 'Berlin', 'Berlin'],
        'Temperature': [5, 7, 3, 8, 2, 4],
        'Humidity': [80, 85, 75, 82, 90, 88]}
df = pd.DataFrame(data)

# Creating a pivot table
pivot_table = pd.pivot_table(df, values=['Temperature', 'Humidity'], index='Date', columns='City', aggfunc=np.mean)

print(pivot_table)
```

    In above example, the pivot_table function takes the DataFrame df and reshapes it. The 'Date' column is set as the index, the 'City' column values become the new columns, and the 'Temperature' and 'Humidity' columns are aggregated using the mean function. The resulting pivot table shows the average temperature and humidity for each city on each date

21.  Why is NumPy’s array slicing faster than Python’s list slicing

    NumPy's array slicing is faster than Python's list slicing due to the fundamental differences in how these data structures are stored and handled in memory. NumPy arrays are stored in contiguous blocks of memory, and they are homogeneous, meaning all elements are of the same data type. This allows for efficient access and manipulation of data, especially when slicing. In contrast, Python lists are stored as arrays of pointers to objects, which can be scattered in memory and can be of different data types.
    When slicing a NumPy array, it can directly compute the memory offset to the desired section of the array and create a new view without copying the underlying data. This operation is very fast because it only involves pointer arithmetic and does not require iterating through elements.
    Python list slicing, however, needs to create a new list object and copy the elements from the original list to the new list. This involves iterating through the elements, allocating memory for the new list, and copying the data, making it a slower process.
    Furthermore, NumPy's operations are often implemented in C and optimized for performance, leveraging vectorization and other techniques to speed up computations. Python lists, on the other hand, rely on Python's interpreted execution model, which is generally slower for numerical operations

```python
# 4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the iris dataset
df = sns.load_dataset('iris')

# Compute the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Set the figure size
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(corr_matrix,
            annot=True,        # Display correlation coefficients
            cmap='coolwarm',   # Color map
            vmin=-1, vmax=1,   # Value range
            square=True,       # Square cells
            linewidths=0.5,    # Lines between cells
            cbar_kws={"shrink": 0.75})  # Color bar size

# Add a title
plt.title('Correlation Heatmap of Iris Dataset', fontsize=16)

# Display the plot
plt.show()
```
